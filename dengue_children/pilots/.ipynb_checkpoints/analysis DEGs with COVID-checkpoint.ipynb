```python
def pair_comparison(adata):    
    cell_types = ['Monocytes', 'Plasmablasts', 'B_cells', 'T_cells', 'NK_cells', 'cDCs', 'pDCs']

    ress = pd.DataFrame([])
    log2FC = pd.DataFrame([])
    
    if 'S_dengue' in adata.obs['Condition'].unique():
        cds = ['dengue', 'S_dengue']
    elif 'Severe' in adata.obs['Condition'].unique():
        cds = ['Moderate', 'Severe']
    
    if 'donor' in adata.obs.columns:
        p_ID = 'donor'
    elif 'ID' in adata.obs.columns:
        p_ID = 'ID'
        
    adata_cd = {cd: (adata[adata.obs['Condition'] == cd])[:, genes] for cd in cds}

    for cell_type in cell_types:
        adata_S_ct = adata_cd[cds[1]][adata_cd[cds[1]].obs['cell_type'] == cell_type]
        adata_NS_ct = adata_cd[cds[0]][adata_cd[cds[0]].obs['cell_type'] == cell_type]
        
        IDs_S = list(adata_S_ct.obs[p_ID].astype('category').cat.categories)
        IDs_NS = list(adata_NS_ct.obs[p_ID].astype('category').cat.categories)

        log2_fc = []
        for ID_S in IDs_S:
            adata_S_ID = adata_S_ct[adata_S_ct.obs[p_ID] == ID_S]
            if len(adata_S_ID.obs_names) < 5:
                continue
            for ID_NS in IDs_NS:
                adata_NS_ID = adata_NS_ct[adata_NS_ct.obs[p_ID] == ID_NS]
                if len(adata_NS_ID.obs_names) < 5:
                    continue

                X_S = adata_S_ID.X
                X_NS = adata_NS_ID.X
                avg_S = np.asarray(X_S.mean(axis=0))[0]
                avg_NS = np.asarray(X_NS.mean(axis=0))[0]
                log2_fc.append(np.log2(avg_S + 0.1) - np.log2(avg_NS + 0.1))

        log2_fc = np.asarray(log2_fc)
        if log2_fc.size == 0:
            continue
        med_pair = np.median(log2_fc, axis=0)
        fra_pair = [len(log2_fc[:, i][log2_fc[:, i] > 0])/log2_fc.shape[0] for i in range(log2_fc.shape[1])]

        res = pd.DataFrame([], index=adata_S_ct.var_names)
        res['med_pair'] = med_pair
        res['fra_pair'] = fra_pair
        res['cell_type'] = cell_type
        ress = pd.concat([ress, res])

        FCs = pd.DataFrame(log2_fc, columns = adata_S_ct.var_names, index=[cell_type] * log2_fc.shape[0])
        log2FC = pd.concat([log2FC, FCs])
        
    return {'pair_res': ress, 'log_FCs': log2FC}
```

```python
def fra_avg(adata):    
    cell_types = ['Monocytes', 'Plasmablasts', 'B_cells', 'T_cells', 'NK_cells', 'cDCs', 'pDCs']

    fra = pd.DataFrame([])
    avg = pd.DataFrame([])

    if 'S_dengue' in adata.obs['Condition'].unique():
        cds = ['dengue', 'S_dengue']
    elif 'Severe' in adata.obs['Condition'].unique():
        cds = ['Moderate', 'Severe']

    if 'donor' in adata.obs.columns:
        p_ID = 'donor'
    elif 'ID' in adata.obs.columns:
        p_ID = 'ID'

    adata_cd = {cd: (adata[adata.obs['Condition'] == cd])[:, genes] for cd in cds}

    for cell_type in cell_types:
        for cd in cds:
            if cell_type not in adata_cd[cd].obs['cell_type'].unique():
                continue
            adata_ct = adata_cd[cd][adata_cd[cd].obs['cell_type'] == cell_type]
            IDs = list(adata_ct.obs[p_ID].astype('category').cat.categories)

            log2_fc = []
            for ID in IDs:
                adata_ID = adata_ct[adata_ct.obs[p_ID] == ID]
                if len(adata_ID.obs_names) < 5:
                    IDs.remove(ID)  
            adata_ct = adata_ct[adata_ct.obs[p_ID].isin(IDs)]
            
            if adata_ct.obs.shape[0] == 0:
                continue
            fra_ct = np.asarray((adata_ct.X > 0).mean(axis=0))[0]
            fra_ct = pd.DataFrame(fra_ct, columns=['fra'], index=adata_ct.var_names)
            fra_ct['cell_type'] = cell_type
            fra_ct['condition'] = cd
            fra = pd.concat([fra, fra_ct])

            avg_ct = np.asarray(adata_ct.X.mean(axis=0))[0]
            avg_ct = pd.DataFrame(avg_ct, columns=['avg'], index=adata_ct.var_names)
            avg_ct['cell_type'] = cell_type
            avg_ct['condition'] = cd
            avg = pd.concat([avg, avg_ct])
    return {'fra': fra, 'avg': avg}
```

```python
def combination(pair, fra, avg):
    if 'dengue' in fra.reset_index()['condition'].unique():
        NS, S = 'dengue', 'S_dengue'
    elif 'Moderate' in fra.reset_index()['condition'].unique():
        NS, S = 'Moderate', 'Severe'
        
    NS_idx = [(i[0], i[1], NS) for i in pair.index]
    S_idx = [(i[0], i[1], S) for i in pair.index]
    
    
    pair['S_fra'] = (fra.loc[S_idx])['fra'].tolist()
    pair['NS_fra'] = (fra.loc[NS_idx])['fra'].tolist()
    
    pair['S_avg'] = (avg.loc[S_idx])['avg'].tolist()
    pair['NS_avg'] = (avg.loc[NS_idx])['avg'].tolist()
    
    return pair
```

```python
'''
filter genes meeting the following criteria:
1. med_pair >= 1 & fra_pair >= 0.9 & S_fra > 2%
2. med_pair <= -1 & fra_pair <= 0.1 & NS_fra > 2%
'''
def filter(D_kids):
    D_kids_f = D_kids[
        ((D_kids['med_pair'] >= 1) & (D_kids['fra_pair'] >= 0.9) & (D_kids['S_fra'] >= 0.02)) |
        ((D_kids['med_pair'] <= -1) & (D_kids['fra_pair'] <= 0.1) & (D_kids['NS_fra'] > 0.02))
    ]
    return D_kids_f
```

```python
def scatter_plot(cell_type, data1, data2, d1_name, d2_name):
    fig, ax = plt.subplots(figsize=[10, 10], dpi=300)
    x = data1.loc[cell_type]['med_pair'].tolist()
    y = data2.loc[cell_type]['med_pair'].tolist()
    ax.scatter(x, y, s=1, c='gray')
    ax.set_xlim(-15, 15)
    ax.set_ylim(-15, 15)
    ax.axvline(0, c='black', zorder=-5, ls='--')
    ax.axhline(0, c='black', zorder=-5, ls='--')
    ax.set_xlabel('Med pair of %s'%d1_name)
    ax.set_ylabel('Med pair of %s'%d2_name)
    ax.set_title(cell_type)
    
    r1 = []
    r2 = []
    r3 = []
    r4 = []
    for i, gene in enumerate(data2.loc[cell_type].index.tolist()):
        ri = ((x[i])**2 + (y[i])**2)**0.5
        if x[i] > 0 and y[i] > 0:
            r1.append([gene, ri])
        elif x[i] < 0 and y[i] > 0:
            r2.append([gene, ri])
        elif x[i] < 0 and y[i] < 0:
            r3.append([gene, ri])
        elif x[i] > 0 and y[i] < 0:
            r4.append([gene, ri])
            
    genes = {}      
    for i, (r, color) in enumerate(zip([r1, r2, r3, r4], ['red', 'blue', 'green', 'orange'])):
        r = pd.DataFrame(r, columns=['gene', 'r']).set_index('gene').sort_values('r')
        genes[i] = r.tail(10).index.tolist()
        mark_x = [data1.loc[cell_type]['med_pair'].loc[gene] for gene in genes[i]]
        mark_y = [data2.loc[cell_type]['med_pair'].loc[gene] for gene in genes[i]]
        for gene, x, y in zip(genes[i], mark_x, mark_y):
            ax.text(x, y, gene, c='black', fontsize=5)
        ax.scatter(mark_x, mark_y, c=color, s=2)

    return {'fig': fig, 'ax': ax, 'genes': genes}
```

```python
def randomization(genes, adata):
    '''
    genes = {
    ct1: [],
    ct2: [],
    ct3: [],
    }
    '''
    log2fc = defaultdict(list)
    pvalue = {}

    if 'S_dengue' in adata.obs['Condition'].unique():
        cds = ['dengue', 'S_dengue']
    elif 'Severe' in adata.obs['Condition'].unique():
        cds = ['Moderate', 'Severe']

    if 'donor' in adata.obs.columns:
        p_ID = 'donor'
    elif 'ID' in adata.obs.columns:
        p_ID = 'ID'

    adata_cd = {cd: (adata[adata.obs['Condition'] == cd]) for cd in cds}

    for cell_type in genes.keys():
        adata_NS = adata_cd[cds[0]][adata_cd[cds[0]].obs['cell_type'] == cell_type]
        adata_S = adata_cd[cds[1]][adata_cd[cds[1]].obs['cell_type'] == cell_type]

        IDs_NS = list(adata_NS.obs[p_ID].astype('category').cat.categories)
        IDs_S = list(adata_S.obs[p_ID].astype('category').cat.categories)
            
        for ID in IDs_NS:
            adata_ID = adata_NS[adata_NS.obs[p_ID] == ID]
            if len(adata_ID.obs_names) < 5:
                IDs_NS.remove(ID)  
        adata_NS = adata_NS[adata_NS.obs[p_ID].isin(IDs_NS)]
        
        for ID in IDs_S:
            adata_ID = adata_S[adata_S.obs[p_ID] == ID]
            if len(adata_ID.obs_names) < 5:
                IDs_S.remove(ID)  
        adata_S = adata_S[adata_S.obs[p_ID].isin(IDs_S)]

        for gene in genes[cell_type]:
            X_NS = adata_NS[:, gene].X.toarray()[:, 0]
            X_S = adata_S[:, gene].X.toarray()[:, 0]
            avg_NS = X_NS.mean()
            avg_S = X_S.mean()
            lfc0 = np.log2(avg_S + 0.1) - np.log2(avg_NS + 0.1)
            log2fc[(cell_type, gene)].append(lfc0)

            p = 0
            X = X_S.tolist() + X_NS.tolist()
            for i in range(1000):
                idx_Si = random.sample(range(len(X)), len(X_S))
                X_Si = [X[i] for i in idx_Si]
                remain_idx = tuple(set(range(len(X))) - set(idx_Si))
                X_NSi = [X[i] for i in remain_idx]
                lfci = np.log2(np.mean(X_Si) + 0.1) - np.log2(np.mean(X_NSi) + 0.1)
                log2fc[(cell_type, gene)].append(lfci)
                if abs(lfci) >= abs(lfc0):
                    p += 1
            pvalue[(cell_type, gene)] = p * 0.001

    res = pd.DataFrame([])
    for key in log2fc.keys():
        log2fc[key] = pd.DataFrame(log2fc[key], columns = ['log2fc'])
        log2fc[key]['pvalue'] = pvalue[key]
        for i, s in enumerate(['cell_type', 'gene']):
            log2fc[key][s] = key[i]
        res = pd.concat([res, log2fc[key]])
    res = res.set_index(['cell_type', 'gene'])
        
    return res
```

```python
def ran_filter(ran_res, genes):
    sig_res = pd.DataFrame([], columns = ['log2fc', 'pvalue', 'cell_type', 'gene'])
    i = 0
    for cell_type in genes.keys():
        for gene in genes[cell_type]:
            if ran_res.loc[cell_type, gene][:1]['pvalue'][0] == 0:
                i += 1
                loc = ran_res.loc[cell_type, gene][:1].loc[cell_type, gene].tolist()
                loc.append(cell_type)
                loc.append(gene)
                sig_res.loc[i] = loc
    sig_res = sig_res.set_index(['cell_type', 'gene'])
    return sig_res
```

```python
print('define a function to find genes whether specific for a data ')
def genes_cat(data1, data2):
    both_genes_idx = [idx for idx in data1.index if idx in data2.index]
    data1_genes_idx = [idx for idx in data1.index if idx not in data2.index]
    data2_genes_idx = [idx for idx in data2.index if idx not in data1.index]
#     both_genes = pd.DataFrame([], 
#                               columns = ['med_pair'+cd1, 'fra_pair'+cd1, 'S_fra'+cd1, 'NS_fra'+cd1, 'S_avg'+cd1, 'NS_avg'+cd1,
#                                          'med_pair'+cd2, 'fra_pair'+cd2, 'S_fra'+cd2, 'NS_fra'+cd2, 'S_avg'+cd2, 'NS_avg'+cd2]
#                              )
    both_genes = pd.concat([data1.loc[both_genes_idx], data2.loc[both_genes_idx]], axis=1)
    data1_genes = data1.loc[data1_genes_idx]
    data2_genes = data2.loc[data2_genes_idx]
    
    return {'both': both_genes, 'data1': data1_genes, 'data2': data2_genes}
```
